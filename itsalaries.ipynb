```python
data = pd.read_csv('IT Salary Survey EU 2020.csv')
data.head()
```

```python
data.info()
```

```python
#salaries statistics
data.describe()
```

```python
data.isnull().sum()
```

```python
data["Gender"].fillna("No Gender", inplace = True)
 
data.isnull().sum() 
```

```python
data = data.dropna(axis = 0, how ='any')

print(data.isnull().sum())
data.shape
```

```python

# plot a histogram
data['Yearly brutto salary (without bonus and stocks) in EUR'].hist(bins=10)

# shows presence of a lot of outliers/extreme values
data.boxplot(column='Yearly brutto salary (without bonus and stocks) in EUR', by = 'Annual brutto salary (without bonus and stocks) one year ago. Only answer if staying in the same country')

# plotting points as a scatter plot
x = data["Yearly brutto salary (without bonus and stocks) in EUR"]
y = data["Annual brutto salary (without bonus and stocks) one year ago. Only answer if staying in the same country"]
plt.scatter(x, y, label= "stars", color= "m",
			marker= "*", s=30)
# x-axis label
plt.xlabel('Yearly brutto salary (without bonus and stocks) in EUR')
# frequency label
plt.ylabel('Annual brutto salary (without bonus and stocks) one year ago. Only answer if staying in the same country')
# function to show the plot
plt.show()

```
